```python
import pandas as pd
from datetime import datetime as dt, timedelta
import plotly.express as px
import plotly.graph_objects as go
import plotly.colors
```

```python
data = pd.read_csv('/content/online_retail.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.dropna(subset=['CustomerID'], inplace=True)
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data['TotalAmount'] = data['Quantity'] * data['UnitPrice']
```

```python
data.head()
```

```python
reference_date = pd.Timestamp(dt.now().date())
```

```python
reference_date = data['InvoiceDate'].max() + timedelta(days=1)
```

```python
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalAmount': 'sum'
})
```

```python
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalAmount': 'Value'
}, inplace=True)
```

```python
rfm.head()
```

```python
# Define quantiles
quantiles = rfm.quantile(q=[0.25, 0.5, 0.75])

#Assign RFM SCORES
def RScore(x, p, d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1

def FScore(x, p, d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4

def MScore(x, p, d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4

rfm['R'] = rfm['Recency'].apply(RScore, args=('Recency', quantiles))
rfm['F'] = rfm['Frequency'].apply(FScore, args=('Frequency', quantiles))
rfm['M'] = rfm['Value'].apply(MScore, args=('Value', quantiles))
```

```python
rfm.head()
```

```python
rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
```

```python
rfm.head()
```

```python
segment_labels =['Low-Value', 'Mid-Value', 'High-Value']

def assign_segment(score):
  if score < 5:
    return 'Low-Value'
  elif score < 9:
    return 'Mid-Value'
  else:
    return 'High-Value'

# Calculate numerical RFM score by summing R, F, and M
rfm['RFM_Score_Numeric'] = rfm['R'] + rfm['F'] + rfm['M']

# Apply the assign_segment function to the numerical score
rfm['RFM_Segment_Label'] = rfm['RFM_Score_Numeric'].apply(assign_segment)

rfm.head()
```

```python
segment_counts = rfm['RFM_Segment_Label'].value_counts().reset_index()
segment_counts.columns = ['RFM_Segment_Label', 'Count']
segment_counts = segment_counts.sort_values(by='RFM_Segment_Label')
```

```python
# Create the bar chart using Plotly

fig = px.bar(segment_counts, x='RFM_Segment_Label', y='Count',
             labels={'RFM_Segment_Label': 'RFM Segment', 'Count': 'Count'},
             title='Distribution of RFM Segments',
             color='RFM_Segment_Label',
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.show()
```

```python
rfm['RFM_Customer_Segments']='Vip/Loyal'

rfm.loc[rfm['RFM_Score_Numeric']>=9,'RFM_Customer_Segments']='Potential Loyal'
rfm.loc[(rfm['RFM_Score_Numeric']>=6) & (rfm['RFM_Score_Numeric']<9),'RFM_Customer_Segments']='AT Risk Customers'
rfm.loc[(rfm['RFM_Score_Numeric']>=5) & (rfm['RFM_Score_Numeric']<5),'RFM_Customer_Segments']="Can't Lose"
rfm.loc[rfm['RFM_Score_Numeric']==4,'RFM_Customer_Segments']="Lost"
segment_counts = rfm.groupby('RFM_Customer_Segments').size().reset_index(name='Count')
segment_product_counts = segment_counts.sort_values('Count',ascending=False)
```

```python
# Create the treemap
fig = px.treemap(segment_product_counts, path=['RFM_Customer_Segments'], values='Count',
                  title='RFM Customer Segments',
                  color='RFM_Customer_Segments',
                 color_discrete_sequence=px.colors.qualitative.Pastel)
#Display the treemap
fig.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Scale the RFM data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Value']])

# Apply the Elbow method to find the optimal number of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()

# Based on the Elbow method, choose an optimal number of clusters (e.g., 4)
# Then apply K-Means clustering with the chosen number of clusters
optimal_k = 4 # Replace with the chosen optimal K from the Elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10) # Added n_init
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Display the first few rows with the new cluster column
display(rfm.head())

# Display the distribution of customers per cluster
display(rfm['Cluster'].value_counts())
```
